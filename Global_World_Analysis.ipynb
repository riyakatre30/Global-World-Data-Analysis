```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Agricultural Land
```

```python
agri = df.groupby('Country')[['AgriculturalLand','LandArea_Km']].sum().nlargest(10,'LandArea_Km')
agri
```

```python
agri = agri / agri.max()

agri.plot(kind='bar', figsize=(10,5))

plt.title("Comparison of Land Area and Agricultural Land")
plt.ylabel("Normalized Value")
plt.xticks(rotation=45)
plt.show()
```

```python
aa = df[['AgriculturalLand','Density_Km','Co2_Emissions']].corr()
aa
```

```python
plt.figure(figsize=(10,5))
sb.heatmap(aa, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
